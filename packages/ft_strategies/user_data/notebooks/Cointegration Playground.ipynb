```python
from cointanalysis import CointAnalysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(400 + np.cumsum(0.1 + np.random.randint(-10, 10, 1000)), columns=['eth'])
df["btc"] = df["eth"] + 50*np.random.random(len(df))
df
```

```python
df.plot()
```

```python
coint = CointAnalysis()
```

```python
coint.test(df[["btc", "eth"]].values).pvalue_
```

```python
coint.fit(df[["btc", "eth"]])
print(f'coef: {coint.coef_}')
print(f'mean: {coint.mean_}')
print(f'std: {coint.std_}')
```

```python
spread = coint.transform(df[["btc", "eth"]].values)
spread
```

```python
spread_ = coint.fit_transform(df[["btc", "eth"]].values)
spread_
```

```python
df["spread"] = spread
```

```python
df[["btc", "eth"]].plot(figsize=(50, 5))
df["spread"].plot(secondary_y=True)
```
