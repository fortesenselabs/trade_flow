```python
import pandas as pd
import numpy as np
```

```python
dataframe = pd.DataFrame(np.random.random(100), columns=["rand"])
dataframe
```

```python
dataframe.isin([np.inf, -np.inf, np.nan]).any()
```

```python
(1 - dataframe["rand"].shift(-1) / dataframe["rand"]).plot()
dataframe["rand"].plot(secondary_y=True)
```

```python
dataframe["pct_c"] = dataframe["rand"].pct_change(2).shift(-2)
dataframe
```
