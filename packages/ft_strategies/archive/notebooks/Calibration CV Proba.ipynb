```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.calibration import CalibratedClassifierCV
from catboost import CatBoostClassifier
import numpy as np
import pandas as pd
```

```python
# Prepare two samples
X, y = make_classification(n_samples=3000, class_sep=0.05, n_informative=30, n_features=100,
                           random_state=0, n_redundant=4, n_clusters_per_class=1)
```

```python
clf = CatBoostClassifier()
```

```python
calibrated_clf = CalibratedClassifierCV(
    base_estimator=clf, method='isotonic', cv=5, n_jobs=5, ensemble=True)
calibrated_clf.fit(X, y)
```

```python
calibrated_clf.calibrated_classifiers_[0]
```

```python
calibrated_clf.predict_proba(X)
```
